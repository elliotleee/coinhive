```python
#simple version of code split

def simplesplit(Target):
    Result = []
    for line in Target:
        templine = line.split(' ')
        if templine[0] in Const:
            templine[0] ='Const'
            templine.append(line)
            Result.append(templine)
        elif templine[0] in Xor:
            templine[0] ='Xor'
            templine.append(line)
            Result.append(templine)
        elif templine[0] in ShiftRow:
            templine[0] ='ShiftRow'
            templine.append(line)
            Result.append(templine)
        elif templine[0] in And:
            templine[0] ='And'
            templine.append(line)
            Result.append(templine)
        elif templine[0] in Mem_operater:
            templine[0] ='Mem_operater'
            templine.append(line)
            Result.append(templine)
    return Result
```
