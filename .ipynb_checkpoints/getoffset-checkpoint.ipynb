```python
file = open("./dataseg.c")
lines= []
```

```python
for line in file.readlines(): 
    #deal with line, delete no need char
    line = line.strip('\n')  
    line = line.strip(' ')
    line = line.strip(';')
    line = line.strip('u')
    line = line.strip(',')
    
    # split line
   
    lines.append(line)
    print(line)
```

```python
lines = [temp for temp in lines if temp != '}' ]
```
